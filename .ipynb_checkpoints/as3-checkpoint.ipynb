```python
from mlxtend.frequent_patterns import apriori
import re
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

```python
final_list, data, stopwords, word_list = [], [], [], []
stopwords = open('stopwords').read()
initdata = open('knock.txt')
dataline = initdata.readlines()

for m in dataline:
    data.append(re.sub('[^A-Za-z0-9]+', ' ', m))

for k in data:
    word = []
    for m in k.split():
        if m.lower() not in stopwords:
            word.append(m.lower())
    final_list.append(word)
dataset = list(filter(None, final_list)) # fastest
```

```python
print(dataset)
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns = te.columns_)
frequent_itemsets = apriori(df, min_support=0.020, use_colnames=True)
```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.50)
```

```python
rules.drop(['antecedent support', 'consequent support', 'support', 'lift', 'leverage', 'conviction'], axis = 1)
```
