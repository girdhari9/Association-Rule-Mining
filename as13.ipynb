```python
# -*- coding: utf-8 -*- 
import re
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
word_list = []
stopwords = []
o = open('stopwords')
stopwords = o.read()
k = open('knock.txt')
s = k.readlines()
data = []
for m in s:
    data.append(re.sub('[^A-Za-z0-9]+', ' ', m))

final_list = []
for k in data:
    word = []
    for m in k.split():
        if m.lower() not in stopwords:
            word.append(m.lower())
    final_list.append(word)
dataset = list(filter(None, final_list)) # fastest
print (dataset)

```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.018, use_colnames=True)
#Since 0.017 was generic/most of the common support.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="confidence", min_threshold=0.75)[['antecedants','consequents','confidence']]
```
